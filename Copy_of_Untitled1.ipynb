```python
# ================================================
# ⚽ Football Team Training Optimization Algorithm
# ================================================

import numpy as np

# ⚙️ إعدادات عامة
POP_SIZE = 50        # عدد الأفراد
MAX_GEN = 100        # عدد التكرارات (الأجيال)
TEAM_SIZE = 5        # حجم كل فريق
```

```python
# 🎯 دالة تقييم الملاءمة (Fitness)
def fitness(individual):
    # الهدف: تقليل الفارق بين العناصر و 0.5
    return -np.sum((individual - 0.5) ** 2)

# 🧬 تهيئة السكان
def initialize_population():
    return np.random.rand(POP_SIZE, 10)  # كل فرد عبارة عن 10 قيم عشوائية
```

```python
# 🏆 إيجاد الأفضل والأسوأ
def get_best_and_worst(pop, fitness_vals):
    best_idx = np.argmax(fitness_vals)
    worst_idx = np.argmin(fitness_vals)
    return best_idx, worst_idx

# 🤝 التدريب الجماعي (Collective Learning)
def collective_training(pop):
    mean = np.mean(pop, axis=0)
    return pop + 0.1 * (mean - pop)

# 👥 تقسيم المجموعات (MGEM)
def clustering_and_grouping(pop):
    np.random.shuffle(pop)
    return [pop[i:i + TEAM_SIZE] for i in range(0, len(pop), TEAM_SIZE)]
```

```python
# 🧪 تدريب كل مجموعة
def group_training(groups):
    trained = []
    for group in groups:
        group_mean = np.mean(group, axis=0)
        group += 0.05 * (group_mean - group)
        trained.append(group)
    return np.vstack(trained)

# 🔁 تحديث اللاعبين
def update_players(pop):
    return pop + np.random.normal(0, 0.01, pop.shape)

# 🧠 تدريب إضافي لأفضل لاعب
def individual_extra_training(pop, best_idx):
    pop[best_idx] += np.random.normal(0, 0.05, pop[best_idx].shape)
    return pop

# 🔄 تحديث السكان
def update_population(old_pop, new_pop, old_fit, new_fit):
    mask = new_fit > old_fit
    return np.where(mask[:, None], new_pop, old_pop)
```

```python
# 🚀 تنفيذ الخوارزمية
def football_team_training():
    population = initialize_population()
    fitness_values = np.array([fitness(ind) for ind in population])

    for gen in range(MAX_GEN):
        best_idx, _ = get_best_and_worst(population, fitness_values)

        population = collective_training(population)
        groups = clustering_and_grouping(population)
        new_population = group_training(groups)

        new_population = update_players(new_population)
        new_population = individual_extra_training(new_population, best_idx)

        new_fitness = np.array([fitness(ind) for ind in new_population])
        population = update_population(population, new_population, fitness_values, new_fitness)
        fitness_values = np.array([fitness(ind) for ind in population])

        if gen % 10 == 0:
            print(f"✅ Generation {gen} | Best fitness: {np.max(fitness_values):.4f}")

    final_best_idx = np.argmax(fitness_values)
    return population[final_best_idx], fitness_values[final_best_idx]
```

```python
# ▶️ تشغيل الخوارزمية
best_solution, best_score = football_team_training()

print("\n🏁 أفضل حل نهائي:", best_solution)
print("💯 أفضل قيمة Fitness:", best_score)
```
